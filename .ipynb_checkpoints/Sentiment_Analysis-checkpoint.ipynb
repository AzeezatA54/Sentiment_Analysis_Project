## Preprocess Tweet Text

```python
nltk.download('stopwords')
nltk.download('punkt_tab')
nltk.download('wordnet')



def preprocess_tweet(text):
    """Cleans the text data and converts it to tokens ready for sentiment analysis.
    
    Args:
        text (str): The tweet text to be cleaned.
    
    Returns:
        str: The cleaned text.
    """
    #Remove URLs, Hashtags
    text = re.sub(r'http\S+|www.\S+|#\w+_', '', text)
    
    
    #Convert text to lower case and tokenize
    tokens = word_tokenize(text.lower())
    stopword = set(stopwords.words('english'))
    tokens = [t for t in tokens if t not in stopword]
    
    #lemmatize text
    lemmatizer = WordNetLemmatizer()
    tokens = [lemmatizer.lemmatize(t) for t in tokens]    
    
    cleaned_text =' '.join(tokens)
    return cleaned_text       
            

Airline2['cleaned_text'] = Airline2['text'].apply(preprocess_tweet)

Airline2.head()
```

### Comparing My Sentiment Analysis with the Dataset's

```python
Sentiment = SentimentIntensityAnalyzer()

sentiment_scores = []
for text in Airline2['cleaned_text']:
    score = Sentiment.polarity_scores(text)
    sentiment_scores.append(score['compound'])
Airline2['new_scores'] = sentiment_scores
Airline2.head()
```

```python
sentiment = []
for score in Airline2['new_scores']:
    if score > 0.5:
        sentiments = 'positive'
    else:
        sentiments = 'negative'
    sentiment.append(sentiments)

Airline2['sentiment'] = sentiment
```
